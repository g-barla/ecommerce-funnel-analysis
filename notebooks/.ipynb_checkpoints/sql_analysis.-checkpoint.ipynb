```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.float_format', '{:.2f}'.format)

# Load CSV
df = pd.read_csv('../data/ecommerce_funnel_data.csv')

print("‚úÖ Data loaded successfully!")
print(f"üìä Shape: {df.shape}")
print(f"\nFirst 3 rows:")
print(df.head(3))
```

```python
# Create SQLite database
db_path = '../data/ecommerce_funnel.db'
conn = sqlite3.connect(db_path)

# Load data into SQL table
df.to_sql('funnel', conn, if_exists='replace', index=False)

print(f"‚úÖ Database created: {db_path}")
print(f"‚úÖ Table 'funnel' created with {len(df):,} rows")

# Verify table creation
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) FROM funnel")
count = cursor.fetchone()[0]
print(f"‚úÖ Verification: {count:,} rows in database")
```

```python
def run_query(query, description=""):
    """
    Execute SQL query and display results nicely
    """
    if description:
        print(f"\n{'='*70}")
        print(f"üîç {description}")
        print('='*70)
    
    result = pd.read_sql_query(query, conn)
    print(result.to_string(index=False))
    print(f"\nüìä Rows returned: {len(result)}")
    
    return result

print("‚úÖ Helper function defined")
```

```python
query1 = """
SELECT 
    COUNT(*) as total_sessions,
    SUM(browsed) as browsed,
    SUM(viewed_product) as viewed_product,
    SUM(added_to_cart) as added_to_cart,
    SUM(reached_checkout) as reached_checkout,
    SUM(completed_purchase) as completed_purchase,
    
    -- Conversion rates at each stage
    ROUND(SUM(viewed_product) * 100.0 / COUNT(*), 2) as view_rate,
    ROUND(SUM(added_to_cart) * 100.0 / COUNT(*), 2) as cart_rate,
    ROUND(SUM(reached_checkout) * 100.0 / COUNT(*), 2) as checkout_rate,
    ROUND(SUM(completed_purchase) * 100.0 / COUNT(*), 2) as purchase_rate,
    
    -- Key metrics
    ROUND(SUM(completed_purchase) * 100.0 / COUNT(*), 2) as overall_conversion,
    ROUND((SUM(added_to_cart) - SUM(reached_checkout)) * 100.0 / 
          NULLIF(SUM(added_to_cart), 0), 2) as cart_abandonment_rate
FROM funnel;
"""

result1 = run_query(query1, "QUERY 1: Overall Funnel Performance")


result1.to_csv('../sql/query1_overall_funnel.csv', index=False)
```

```python
print("\n KEY INSIGHTS - QUERY 1:")
print("-" * 70)

overall_conv = result1['overall_conversion'].values[0]
abandonment = result1['cart_abandonment_rate'].values[0]

print(f"""
1. Overall Conversion: {overall_conv}%
   ‚Üí Out of 100 visitors, {int(overall_conv)} make a purchase
   ‚Üí This is {'EXCELLENT' if overall_conv > 15 else 'GOOD' if overall_conv > 10 else 'NEEDS IMPROVEMENT'}

2. Cart Abandonment: {abandonment}%
   ‚Üí Nearly HALF of cart additions don't convert
   ‚Üí This is the PRIMARY optimization opportunity
   
3. Funnel Drop-offs:
   ‚Üí Browse to View: {100 - result1['view_rate'].values[0]:.1f}% drop (normal)
   ‚Üí View to Cart: {result1['view_rate'].values[0] - result1['cart_rate'].values[0]:.1f}% drop (expected)
   ‚Üí Cart to Checkout: {abandonment:.1f}% drop (CRITICAL ISSUE) ‚ö†Ô∏è
   ‚Üí Checkout to Purchase: {result1['checkout_rate'].values[0] - result1['purchase_rate'].values[0]:.1f}% drop (acceptable)

üìä BUSINESS RECOMMENDATION:
   Focus optimization efforts on cart abandonment - this is where we lose the most revenue.
""")
```

```python
query2 = """
SELECT 
    device,
    COUNT(*) as total_sessions,
    SUM(completed_purchase) as purchases,
    ROUND(SUM(completed_purchase) * 100.0 / COUNT(*), 2) as conversion_rate,
    
    -- Cart abandonment by device
    SUM(added_to_cart) as carts_added,
    SUM(reached_checkout) as checkouts_reached,
    ROUND((SUM(added_to_cart) - SUM(reached_checkout)) * 100.0 / 
          NULLIF(SUM(added_to_cart), 0), 2) as abandonment_rate,
    
    -- Average metrics
    ROUND(AVG(cart_value), 2) as avg_cart_value,
    ROUND(AVG(time_on_page_seconds), 1) as avg_time_on_page
    
FROM funnel
GROUP BY device
ORDER BY conversion_rate DESC;
"""

result2 = run_query(query2, "QUERY 2: Device Performance Analysis")
result2.to_csv('../sql/query2_device_analysis.csv', index=False)
```

```python
print("\nüí° KEY INSIGHTS - QUERY 2:")
print("-" * 70)

best_device = result2.iloc[0]
worst_device = result2.iloc[-1]

print(f"""
DEVICE COMPARISON:

Best Performer: {best_device['device']}
   ‚Üí Conversion: {best_device['conversion_rate']}%
   ‚Üí Abandonment: {best_device['abandonment_rate']}%
   ‚Üí Avg Cart Value: ${best_device['avg_cart_value']}

Worst Performer: {worst_device['device']}
   ‚Üí Conversion: {worst_device['conversion_rate']}%
   ‚Üí Abandonment: {worst_device['abandonment_rate']}%
   ‚Üí Avg Cart Value: ${worst_device['avg_cart_value']}

Gap: {best_device['conversion_rate'] - worst_device['conversion_rate']:.1f} percentage points difference

üìä BUSINESS RECOMMENDATION:
   {'Mobile' if 'Mobile' in worst_device['device'] else worst_device['device']} checkout experience needs optimization.
   Consider: larger buttons, simplified forms, mobile-optimized payment flow.
""")

# Visualization
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(result2['device'], result2['conversion_rate'], color=['#2ecc71', '#3498db', '#e74c3c'])
plt.title('Conversion Rate by Device', fontsize=14, fontweight='bold')
plt.ylabel('Conversion Rate (%)')
plt.ylim(0, max(result2['conversion_rate']) * 1.2)
for i, v in enumerate(result2['conversion_rate']):
    plt.text(i, v + 0.5, f"{v}%", ha='center', fontweight='bold')

plt.subplot(1, 2, 2)
plt.bar(result2['device'], result2['abandonment_rate'], color=['#e74c3c', '#e67e22', '#f39c12'])
plt.title('Cart Abandonment by Device', fontsize=14, fontweight='bold')
plt.ylabel('Abandonment Rate (%)')
plt.ylim(0, max(result2['abandonment_rate']) * 1.2)
for i, v in enumerate(result2['abandonment_rate']):
    plt.text(i, v + 0.5, f"{v}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../visualizations/device_performance.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úÖ Visualization saved: visualizations/device_performance.png")
```

```python
query3 = """
SELECT 
    CASE 
        WHEN shipping_cost = 0 THEN 'Free Shipping'
        WHEN shipping_cost <= 10 THEN 'Low ($1-$10)'
        WHEN shipping_cost <= 15 THEN 'Medium ($11-$15)'
        ELSE 'High (>$15)'
    END as shipping_tier,
    
    COUNT(*) as sessions,
    SUM(added_to_cart) as carts_added,
    SUM(reached_checkout) as checkouts,
    SUM(completed_purchase) as purchases,
    
    ROUND(SUM(completed_purchase) * 100.0 / NULLIF(SUM(added_to_cart), 0), 2) as cart_to_purchase_rate,
    ROUND((SUM(added_to_cart) - SUM(reached_checkout)) * 100.0 / 
          NULLIF(SUM(added_to_cart), 0), 2) as abandonment_rate,
    
    ROUND(AVG(shipping_cost), 2) as avg_shipping_cost,
    ROUND(AVG(cart_value), 2) as avg_cart_value
    
FROM funnel
WHERE added_to_cart = 1  -- Only look at users who added to cart
GROUP BY shipping_tier
ORDER BY 
    CASE shipping_tier
        WHEN 'Free Shipping' THEN 1
        WHEN 'Low ($1-$10)' THEN 2
        WHEN 'Medium ($11-$15)' THEN 3
        WHEN 'High (>$15)' THEN 4
    END;
"""

result3 = run_query(query3, "QUERY 3: Shipping Cost Impact on Abandonment")
result3.to_csv('../sql/query3_shipping_impact.csv', index=False)
```

```python
print("\nüí° KEY INSIGHTS - QUERY 3:")
print("-" * 70)

free_ship = result3[result3['shipping_tier'] == 'Free Shipping'].iloc[0] if 'Free Shipping' in result3['shipping_tier'].values else None
high_ship = result3[result3['shipping_tier'] == 'High (>$15)'].iloc[0]

if free_ship is not None:
    print(f"""
SHIPPING COST IS A CRITICAL ABANDONMENT DRIVER:

Free Shipping:
   ‚Üí Abandonment: {free_ship['abandonment_rate']}%
   ‚Üí Cart-to-Purchase: {free_ship['cart_to_purchase_rate']}%

High Shipping (>$15):
   ‚Üí Abandonment: {high_ship['abandonment_rate']}%
   ‚Üí Cart-to-Purchase: {high_ship['cart_to_purchase_rate']}%

Impact: {high_ship['abandonment_rate'] - free_ship['abandonment_rate']:.1f} percentage point increase in abandonment!

üìä REVENUE CALCULATION:
   High shipping sessions: {high_ship['sessions']:,}
   Abandoned: {int(high_ship['sessions'] * high_ship['abandonment_rate']/100):,}
   Avg cart value: ${high_ship['avg_cart_value']}
   
   Lost revenue: ${int(high_ship['sessions'] * high_ship['abandonment_rate']/100 * high_ship['avg_cart_value']):,}

üìä BUSINESS RECOMMENDATION:
   1. Offer free shipping threshold (e.g., "Free shipping on orders >$75")
   2. Test shipping cost reduction on high-value carts
   3. Hide shipping cost until later in funnel to reduce early abandonment
""")
```

```python
query4 = """
SELECT 
    traffic_source,
    COUNT(*) as total_sessions,
    SUM(completed_purchase) as purchases,
    ROUND(SUM(completed_purchase) * 100.0 / COUNT(*), 2) as conversion_rate,
    
    -- Engagement metrics
    ROUND(AVG(products_viewed), 1) as avg_products_viewed,
    ROUND(AVG(time_on_page_seconds), 0) as avg_time_seconds,
    
    -- Financial metrics
    ROUND(AVG(CASE WHEN completed_purchase = 1 THEN cart_value END), 2) as avg_order_value,
    ROUND(SUM(CASE WHEN completed_purchase = 1 THEN cart_value ELSE 0 END), 2) as total_revenue,
    
    -- Cart abandonment
    ROUND((SUM(added_to_cart) - SUM(reached_checkout)) * 100.0 / 
          NULLIF(SUM(added_to_cart), 0), 2) as abandonment_rate
    
FROM funnel
GROUP BY traffic_source
ORDER BY conversion_rate DESC;
"""

result4 = run_query(query4, "QUERY 4: Traffic Source Performance & ROI")
result4.to_csv('../sql/query4_traffic_source.csv', index=False)
```

```python
print("\nüí° KEY INSIGHTS - QUERY 4:")
print("-" * 70)

best_source = result4.iloc[0]
worst_source = result4.iloc[-1]

# Calculate revenue per session
result4['revenue_per_session'] = result4['total_revenue'] / result4['total_sessions']

print(f"""
TRAFFIC SOURCE COMPARISON:

üèÜ Best Performer: {best_source['traffic_source']}
   ‚Üí Conversion: {best_source['conversion_rate']}%
   ‚Üí Avg Order Value: ${best_source['avg_order_value']}
   ‚Üí Total Revenue: ${best_source['total_revenue']:,.2f}
   ‚Üí Engagement: {best_source['avg_products_viewed']} products, {int(best_source['avg_time_seconds'])}s

üìâ Worst Performer: {worst_source['traffic_source']}
   ‚Üí Conversion: {worst_source['conversion_rate']}%
   ‚Üí Avg Order Value: ${worst_source['avg_order_value']}
   ‚Üí Total Revenue: ${worst_source['total_revenue']:,.2f}

Performance Gap: {best_source['conversion_rate'] - worst_source['conversion_rate']:.1f} percentage points

üìä REVENUE PER SESSION:
""")

for _, row in result4.iterrows():
    rev_per_session = row['total_revenue'] / row['total_sessions']
    print(f"   {row['traffic_source']:20s} ‚Üí ${rev_per_session:.2f}/session")

print(f"""
üìä BUSINESS RECOMMENDATION:
   1. Increase budget for {best_source['traffic_source']} (highest ROI)
   2. Investigate why {worst_source['traffic_source']} underperforms
   3. A/B test landing pages by traffic source
   4. Consider reducing spend on low-converting sources
""")
```

```python
query5 = """
SELECT 
    CASE WHEN has_discount = 1 THEN 'With Discount' ELSE 'No Discount' END as discount_status,
    
    COUNT(*) as total_sessions,
    SUM(added_to_cart) as carts_added,
    SUM(completed_purchase) as purchases,
    
    -- Conversion metrics
    ROUND(SUM(completed_purchase) * 100.0 / COUNT(*), 2) as overall_conversion,
    ROUND(SUM(completed_purchase) * 100.0 / NULLIF(SUM(added_to_cart), 0), 2) as cart_to_purchase,
    ROUND((SUM(added_to_cart) - SUM(reached_checkout)) * 100.0 / 
          NULLIF(SUM(added_to_cart), 0), 2) as abandonment_rate,
    
    -- Financial impact
    ROUND(AVG(cart_value), 2) as avg_cart_value,
    ROUND(SUM(CASE WHEN completed_purchase = 1 THEN cart_value ELSE 0 END), 2) as total_revenue,
    ROUND(AVG(CASE WHEN completed_purchase = 1 THEN cart_value END), 2) as avg_order_value
    
FROM funnel
GROUP BY has_discount
ORDER BY discount_status DESC;
"""

result5 = run_query(query5, "QUERY 5: Discount Impact on Conversion & Abandonment")
result5.to_csv('../sql/query5_discount_impact.csv', index=False)
```

```python
print("\nüí° KEY INSIGHTS - QUERY 5:")
print("-" * 70)

with_discount = result5[result5['discount_status'] == 'With Discount'].iloc[0]
no_discount = result5[result5['discount_status'] == 'No Discount'].iloc[0]

conv_lift = with_discount['overall_conversion'] - no_discount['overall_conversion']
abandonment_reduction = no_discount['abandonment_rate'] - with_discount['abandonment_rate']

print(f"""
DISCOUNT EFFECTIVENESS:

With Discount:
   ‚Üí Conversion: {with_discount['overall_conversion']}%
   ‚Üí Abandonment: {with_discount['abandonment_rate']}%
   ‚Üí Avg Order Value: ${with_discount['avg_order_value']}
   ‚Üí Total Revenue: ${with_discount['total_revenue']:,.2f}

Without Discount:
   ‚Üí Conversion: {no_discount['overall_conversion']}%
   ‚Üí Abandonment: {no_discount['abandonment_rate']}%
   ‚Üí Avg Order Value: ${no_discount['avg_order_value']}
   ‚Üí Total Revenue: ${no_discount['total_revenue']:,.2f}

üìà DISCOUNT IMPACT:
   ‚Üí Conversion Lift: +{conv_lift:.1f} percentage points
   ‚Üí Abandonment Reduction: -{abandonment_reduction:.1f} percentage points
   ‚Üí Relative improvement: {(conv_lift / no_discount['overall_conversion'] * 100):.1f}% better conversion

üí∞ ROI CALCULATION:
   Sessions with discount: {with_discount['total_sessions']:,}
   Extra purchases due to discount: ~{int(with_discount['total_sessions'] * conv_lift / 100):,}
   Revenue from those purchases: ${int(with_discount['total_sessions'] * conv_lift / 100 * with_discount['avg_order_value']):,}
   
üìä BUSINESS RECOMMENDATION:
   Discounts {'ARE' if conv_lift > 3 else 'MIGHT BE'} worth the margin reduction:
   1. Test targeted discounts for cart abandoners
   2. Implement "first purchase" discount codes
   3. Use exit-intent popups with discount offers
   4. A/B test discount percentage (10% vs 15% vs 20%)
""")

# Comparison visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Conversion comparison
axes[0].bar(['With Discount', 'No Discount'], 
            [with_discount['overall_conversion'], no_discount['overall_conversion']],
            color=['#27ae60', '#95a5a6'])
axes[0].set_title('Overall Conversion Rate', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Conversion Rate (%)')
for i, v in enumerate([with_discount['overall_conversion'], no_discount['overall_conversion']]):
    axes[0].text(i, v + 0.5, f"{v}%", ha='center', fontweight='bold')

# Abandonment comparison
axes[1].bar(['With Discount', 'No Discount'], 
            [with_discount['abandonment_rate'], no_discount['abandonment_rate']],
            color=['#27ae60', '#e74c3c'])
axes[1].set_title('Cart Abandonment Rate', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Abandonment Rate (%)')
for i, v in enumerate([with_discount['abandonment_rate'], no_discount['abandonment_rate']]):
    axes[1].text(i, v + 0.5, f"{v}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../visualizations/discount_impact.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úÖ Visualization saved: visualizations/discount_impact.png")
```
